```python
from math import nan
import pandas as pd
import json, os, re

data = []
root = "./raw_data/"
for file in os.listdir(root):
    data.append(json.load(open(root+file)))

data = pd.DataFrame(data)
for col in data.columns:
    data[col] = data[col].apply(lambda x: nan if type(x) != str and x == None else x )

data
```
