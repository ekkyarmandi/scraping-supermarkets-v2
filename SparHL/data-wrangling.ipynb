```python
from math import nan
import pandas as pd
import json, os, re

data = []
root = "./raw_data/"
for file in os.listdir(root):
    data.append(json.load(open(root+file)))

data = pd.DataFrame(data)
for col in data.columns:
    data[col] = data[col].apply(lambda x: nan if type(x) != str and x == None else x)
    
for i in data.index:
    
    # update single nutriscore
    if i == 5802:
        data.loc[i,"nutriscore"] = "A"
    
    # remove ingredients word
    word = "ingrediënten"
    text = data.loc[i,'ingredients_text']
    if type(text) == str and word in text.lower():
        a,b = re.search(word,text.lower()).span()
        text = text.replace(text[a:b],"").strip(":").strip()
        data.loc[i,'ingredients_text'] = text
        
    # data.loc[i,"sources"] = [data.loc[i,"sources"]]

data = data.drop(columns="it_has_nutriscore")
data = data.dropna(subset="code")
data
```

```python
product_data = []
for i in data.index:
    product = data.loc[i].to_dict()
    for k,v in product.items():
        if type(v) == float:
            product[k] = None
    product_data.append(product)
json.dump(product_data,open("./data/product_data.json","w"))
print("done")
```
