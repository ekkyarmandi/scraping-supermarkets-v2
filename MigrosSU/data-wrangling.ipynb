```python
from math import nan
import pandas as pd
import json, os, re

data = []
root = "./raw_data/"
for file in os.listdir(root):
    data.append(json.load(open(root+file)))

data = pd.DataFrame(data)
for col in data.columns:
    data[col] = data[col].apply(lambda x: nan if type(x) != str and x == None else x)
    
data = data.drop(columns="it_has_nutriscore")
data = data.dropna(subset="code")

for i in data.index:
    gtin = data.loc[i,'code']
    if len(gtin) == 1:
        data.loc[i,'code'] = gtin[0]
data
```

```python
product_data = []
for i in data.index:
    product = data.loc[i].to_dict()
    for k,v in product.items():
        if type(v) == float:
            product[k] = None
    product_data.append(product)
json.dump(product_data,open("./data/product_data.json","w"))
print("done")
```
